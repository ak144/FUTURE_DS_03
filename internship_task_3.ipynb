Perform sentiment analysis on the text data in the DataFrame, analyze the sentiment by course, and visualize the results.

## Identify text data

### Subtask:
Determine which columns contain text data relevant to sentiment analysis.


**Reasoning**:
Inspect the DataFrame's information to identify columns with object data types, which likely contain text data for sentiment analysis.



```python
display(df.info())
```

**Reasoning**:
Based on the `df.info()` output, the 'Questions', 'Average/ Percentage', 'Course Name', and 'Basic Course' columns have an object data type, indicating they contain text. Among these, 'Questions' is the most likely column to contain textual feedback relevant to sentiment analysis. I will display the unique values of the 'Questions' column to confirm its content.



```python
display(df['Questions'].unique())
```

```python
from textblob import TextBlob

sentiments = []
for text_value in df['Questions']:
        blob = TextBlob(text_value)
        sentiments.append(blob.sentiment.polarity)
df['sentiment'] = sentiments
print(df)
```

```python
def getAnalysis(score):
 if score < 0:
     return 'Negative'
 elif score == 0:
     return 'Neutral'
 else:
    return 'Positive'
df['TextBlob_Analysis'] = df['sentiment'].apply(getAnalysis)
print(df.head(20))
```
